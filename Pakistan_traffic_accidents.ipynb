```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Pakistani_Traffic_Accidents.csv')
```

```python
# storing the data in another data frame , in case I do some unwanted changings
df1 = pd.DataFrame(df)
```

```python
# deleting the first row as it contain NaN values
df = df.drop(df.index[0])
df = df.reset_index(drop=True)
```

```python
# the first row contains column names
col_names = df.iloc[0]  
```

```python
df = df[1:]  
```

```python
df.columns = col_names  
df = df.reset_index(drop=True)
```

```python
# setting the notebook to print all the rows
pd.set_option('display.max_rows',None)
```

```python
df
```

```python
# extracting only the country data into a separate data frame
whole_country_data = df.iloc[:11]
whole_country_data.reset_index(drop=True, inplace=True)
```

```python
# from row 11 to 55 it contain the provincial data, so storing it in a separate data frame
provincial_data = df.iloc[11:55]
provincial_data.reset_index(drop=True, inplace=True)
# setting the column name to "Province"
provincial_data.rename(columns={np.nan: 'Province'}, inplace=True)
```

```python
# the remaining data is for islamabad so storing it separately
islamabad_data =  df.iloc[55:]
islamabad_data.reset_index(drop=True, inplace=True)
```

### In which  province the most accidents have occured?

```python
provincial_data['Total number of accidents'].dtype
```

```python
# converting the data type from object to numeric for mathematical operations
provincial_data['Total number of accidents'] = pd.to_numeric(provincial_data['Total number of accidents'], errors='coerce')
```

```python
# grouping the data by each province and adding total accidents accross all the years
province_total_accidents = provincial_data.groupby('Province')['Total number of accidents'].sum()
```

```python
# finding the province with the highest total number of accidents, it will give the province name
province_with_most_accidents = province_total_accidents.idxmax()
```

```python
# it will give the actual count "no of accidents"
most_accidents_count = province_total_accidents.max()
```

```python
print(f"The province with the most accidents is {province_with_most_accidents} with {most_accidents_count} accidents.")
```

### In which province the least accidents have accured?

```python
# finding the province with the least total number of accidents
province_with_least_accidents = province_total_accidents.idxmin()
```

```python
least_accidents_count = province_total_accidents.min()
```

```python
print(f"The province with the least accidents is {province_with_least_accidents} with {least_accidents_count} accidents.")
```

## Bonus Questions

### Visualizing how the total number of accidents have changed the over the years.

```python
# getting th Year and total no of accidents columns
years = whole_country_data['Year']
accidents = whole_country_data['Total number of accidents']
```

```python
# creating a line plot to visualize the trend of total accidents over the years
plt.figure(figsize=(10, 6))
plt.plot(years, accidents, marker='o', color='b', linestyle='-')
plt.title('Total Number of Accidents Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Number of Accidents')
plt.grid(True)
plt.show()
```

## Accident Statistics Comparison Across Provinces

```python
# grouping by province
province_grouped = provincial_data.groupby('Province')
```

```python
# getting the stats , it will contain all these columns but all summed up for the total years
province_stats = province_grouped[['Total number of accidents', 'Total number of vehicles involved', 'Killed', 'Injured']].sum()
```

```python
# creating a grouped bar plot
province_stats.plot(kind='bar', figsize=(12, 8))
plt.title('Accident Statistics by Province')
plt.xlabel('Province')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Statistics')
plt.tight_layout()
plt.show()
```

## Trend Comparison: Fatal vs. Non-Fatal Accidents Over Time

```python
years = whole_country_data['Year']
fatal_accidents = whole_country_data['Fatal Accidents']
non_fatal_accidents = whole_country_data['Non-Fatal Accidents']
total_vehicles_involved = whole_country_data['Total number of vehicles involved']
```

```python
# creating sub-plots
fig, axes = plt.subplots(2, 1, figsize=(12, 10))

# for fatal Vs non-fatal accidents
axes[0].plot(years, fatal_accidents, marker='o', label='Fatal Accidents', color='red')
axes[0].plot(years, non_fatal_accidents, marker='o', label='Non-Fatal Accidents', color='blue')
axes[0].set_title('Fatal vs. Non-Fatal Accidents Over the Years')
axes[0].set_xlabel('Year')
axes[0].set_ylabel('Number of Accidents')
axes[0].legend()
axes[0].grid(True)

# plotting for total accidents involved
axes[1].plot(years, total_vehicles_involved, marker='o', color='green')
axes[1].set_title('Total Vehicles Involved in Accidents Over the Years')
axes[1].set_xlabel('Year')
axes[1].set_ylabel('Total Vehicles Involved')
axes[1].grid(True)
```
